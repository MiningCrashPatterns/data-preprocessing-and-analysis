```python
drop_redundant_accident = [
    "STATE",        # duplicate of STATENAME
    "COUNTY",       # duplicate of COUNTYNAME
    "CITY",         # duplicate of CITYNAME
    "MONTH",        # duplicate of MONTHNAME
    "DAY",          # duplicate of DAYNAME
    "DAY_WEEK",     # duplicate of DAY_WEEKNAME
    "HOUR",         # duplicate of HOURNAME
    "MINUTE",       # duplicate of MINUTENAME
    "ROUTE",        # duplicate of ROUTENAME
    "RUR_URB",      # duplicate of RUR_URBNAME
    "FUNC_SYS",     # duplicate of FUNC_SYSNAME
    "RD_OWNER",     # duplicate of RD_OWNERNAME
    "NHS",          # duplicate of NHSNAME
    "SP_JUR",       # duplicate of SP_JURNAME
    "MILEPT",       # duplicate of MILEPTNAME
    "LATITUDE",     # duplicate of LATITUDENAME
    "LONGITUD",     # duplicate of LONGITUDNAME
    "HARM_EV",      # duplicate of HARM_EVNAME
    "MAN_COLL",     # duplicate of MAN_COLLNAME
    "RELJCT1",      # duplicate of RELJCT1NAME
    "RELJCT2",      # duplicate of RELJCT2NAME
    "TYP_INT",      # duplicate of TYP_INTNAME
    "REL_ROAD",     # duplicate of REL_ROADNAME
    "WRK_ZONE",     # duplicate of WRK_ZONENAME
    "LGT_COND",     # duplicate of LGT_CONDNAME
    "WEATHER",      # duplicate of WEATHERNAME
    "SCH_BUS",      # duplicate of SCH_BUSNAME
    "RAIL",         # duplicate of RAILNAME
    "NOT_HOUR",     # duplicate of NOT_HOURNAME
    "NOT_MIN",      # duplicate of NOT_MINNAME
    "ARR_HOUR",     # duplicate of ARR_HOURNAME
    "ARR_MIN",      # duplicate of ARR_MINNAME
    "HOSP_HR",      # duplicate of HOSP_HRNAME
    "HOSP_MN",      # duplicate of HOSP_MNNAME
]
```

```python
accidents_df = accidents_df.drop(
    columns=[c for c in drop_redundant_accident if c in accidents_df.columns]
)

print("âœ… Dropped redundant columns successfully.")
print(f"Remaining columns count: {len(accidents_df.columns)}")
print("\nList of columns:\n")
print(accidents_df.columns.tolist())
```

```python
# ------------------------------------------------------------
# ACCIDENT columns (cleaned) - grouped
# ------------------------------------------------------------

# 1) core identifiers / targets  âœ… keep
core_cols = [
    "ST_CASE",     # unique crash ID
    "YEAR",        # crash year
    "FATALS",      # outcome: number of fatalities
]

# 2) counts / severity context  âœ… usually keep
count_cols = [
    "PEDS",        # number of pedestrians involved
    "PERNOTMVIT",  # people not in motor vehicles
    "VE_TOTAL",    # number of vehicles in crash
    "VE_FORMS",    # number of vehicle forms (FARS admin)
    "PVH_INVL",    # parked vehicle involved
    "PERSONS",     # total persons in crash
    "PERMVIT",     # persons in motor vehicles in transport
]

# 3) location (state/city/county)  âœ… usually keep
location_cols = [
    "STATENAME",    # state of crash
    "COUNTYNAME",   # county of crash
    "CITYNAME",     # city/place name
]

# 4) time / calendar  âœ… usually keep
time_cols = [
    "MONTHNAME",     # month of crash
    "DAYNAME",       # day of month
    "DAY_WEEKNAME",  # day of week
    "HOURNAME",      # hour group of crash
    "MINUTENAME",    # minute group of crash
]

# 5) roadway / system descriptors  âœ… useful for EDA
road_env_cols = [
    "ROUTENAME",      # route type (interstate, state routeâ€¦)
    "RUR_URBNAME",    # rural vs urban
    "FUNC_SYSNAME",   # functional class
    "RD_OWNERNAME",   # road ownership
    "NHSNAME",        # on National Highway System?
    "SP_JURNAME",     # special jurisdiction (tribal, federalâ€¦)
]

# 6) crash characteristics  âœ… useful
crash_desc_cols = [
    "HARM_EVNAME",    # first harmful event
    "MAN_COLLNAME",   # manner of collision
    "RELJCT1NAME",    # relation to junction (general)
    "RELJCT2NAME",    # relation to junction (specific)
    "TYP_INTNAME",    # type of intersection
    "REL_ROADNAME",   # relation to roadway
    "WRK_ZONENAME",   # work zone presence/type
    "LGT_CONDNAME",   # lighting condition
    "WEATHERNAME",    # weather condition
    "SCH_BUSNAME",    # school bus involvement
    "RAILNAME",       # rail crossing involvement
]

# 7) operational / response times  ðŸŸ¡ keep only if you study EMS
response_time_cols = [
    "NOT_HOURNAME",  # hour notified
    "NOT_MINNAME",   # minute notified
    "ARR_HOURNAME",  # arrival hour
    "ARR_MINNAME",   # arrival minute
    "HOSP_HRNAME",   # hospital hour
    "HOSP_MNNAME",   # hospital minute
]

# 8) text-ish / highly specific location  ðŸŸ¡ often drop for ML
texty_location_cols = [
    "TWAY_ID",       # primary street name
    "TWAY_ID2",      # cross street name
    "MILEPTNAME",    # milepoint text
    "LATITUDENAME",  # latitude text
    "LONGITUDNAME",  # longitude text
]

# ------------------------------------------------------------
# Now decide what to drop for a lean ML/EDA table
# ------------------------------------------------------------
# Drop these first: they are free-text or very high cardinality
drop_for_model = texty_location_cols

# Optionally drop EMS/response times if not studying timelines
drop_for_model += response_time_cols

# Build final keep list
all_cols = (
    core_cols
    + count_cols
    + location_cols
    + time_cols
    + road_env_cols
    + crash_desc_cols
    + response_time_cols
    + texty_location_cols
)

accidents_clean = accidents_df.drop(
    columns=[c for c in drop_for_model if c in accidents_df.columns]
)

print("Original cols:", accidents_df.shape[1])
print("After cleanup:", accidents_clean.shape[1])
print("Remaining columns:", accidents_clean.columns.tolist())
```
