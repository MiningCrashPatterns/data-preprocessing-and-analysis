```python
# ------------------------------------------------------------
# Vehicle table cleanup
# Goal: drop columns that are too detailed, rarely used, or
#       specific to commercial/hazmat so merges stay lean.
# ------------------------------------------------------------

# 1) VIN-related columns (very granular, usually not needed)
vin_cols = [
    "VINNAME",
    "VIN_1","VIN_2","VIN_3","VIN_4","VIN_5","VIN_6",
    "VIN_7","VIN_8","VIN_9","VIN_10","VIN_11","VIN_12",
]

# 2) Trailer-related columns (only for big trucks, often empty)
trailer_cols = [
    "TRLR1VIN","TRLR1VINNAME",
    "TRLR2VIN","TRLR2VINNAME",
    "TRLR3VIN","TRLR3VINNAME",
    "TRLR1GVWR","TRLR1GVWRNAME",
    "TRLR2GVWR","TRLR2GVWRNAME",
    "TRLR3GVWR","TRLR3GVWRNAME",
]

# 3) Hazmat / cargo columns (not useful for person-level EDA)
hazmat_cols = [
    "CARGO_BTNAME",
    "HAZ_INVNAME","HAZ_PLACNAME","HAZ_IDNAME",
    "HAZ_CNONAME","HAZ_RELNAME",
]

# 4) Commercial carrier / bus / special use (niche)
carrier_cols = [
    "MCARR_ID","MCARR_IDNAME",
    "MCARR_I1","MCARR_I1NAME",
    "MCARR_I2","MCARR_I2NAME",
    "BUS_USENAME",
    "SPEC_USENAME",
    "EMER_USENAME",
]

# 5) Driver/license/ZIP/physicals (admin-level, not vehicle-level)
driver_cols = [
    "L_STATENAME",
    "L_TYPENAME",
    "L_STATUSNAME",
    "CDL_STATNAME",
    "L_ENDORSNAME",
    "L_COMPLNAME",
    "L_RESTRINAME",
    "DR_ZIP","DR_ZIPNAME",
    "DR_HGTNAME","DR_WGTNAME",
    "DR_DRINKNAME",
    "DR_PRESNAME",
]

# 6) Driver history and license dates (nice-to-have, but noisy)
history_cols = [
    "PREV_ACCNAME",
    "PREV_SUS1NAME","PREV_SUS2NAME","PREV_SUS3NAME",
    "PREV_DWINAME",
    "PREV_SPDNAME",
    "PREV_OTHNAME",
    "FIRST_MONAME","FIRST_YRNAME",
    "LAST_MONAME","LAST_YRNAME",
]

# 7) Misc very specific fields we can drop now
misc_cols = [
    "MAK_MOD","MAK_MODNAME",
    "VTCONT_FNAME",
    "UNDEROVERRIDENAME",
    "J_KNIFENAME",
]

# combine all drop lists
cols_to_drop = (
    vin_cols
    + trailer_cols
    + hazmat_cols
    + carrier_cols
    + driver_cols
    + history_cols
    + misc_cols
)

# actually drop them (only if present)
vehicles_clean = vehicles_df.drop(
    columns=[c for c in cols_to_drop if c in vehicles_df.columns]
)

print("Original cols:", vehicles_df.shape[1])
print("After cleanup:", vehicles_clean.shape[1])
print("Remaining columns:", vehicles_clean.columns.tolist())
```
